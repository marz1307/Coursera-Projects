```python
import requests
import json
import pandas as pd
from pandas import json_normalize
data = requests.get("https://official-joke-api.appspot.com/jokes/ten")
result = json.loads(data.text)
pd.DataFrame(result)
df=pd.json_normalize(result)
df
df.drop(columns=["type","id"], inplace=True)
df
```
